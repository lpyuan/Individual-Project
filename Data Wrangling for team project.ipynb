```python
worldcup = pd.read_csv('/Users/lipinyuan/Downloads/DataViz/fifa-world-cup/WorldCups.csv')
```

```python
worldcup = worldcup.rename(columns={'Year': 'Host_Year'})
```

```python
worldcup.head()
```

```python
worldcup_new = worldcup.drop(worldcup.columns[2:11], axis=1)
```

```python
worldcup_new = worldcup_new.replace("USA","United States").replace("England","United Kingdom").replace("Korea/Japan","Korea, Rep.")
worldcup_new = worldcup_new.append({'Host_Year':'2002', 'Country':'Japan'},ignore_index=True)
```

```python
worldcup_new
```

```python
worldcup_new.Host_Year = worldcup_new.Host_Year.astype(np.int)
```

```python
worldcup_new.dtypes
```

# Finding Two Data Wrangling Part：

```python
GDP = pd.read_csv('/Users/lipinyuan/Downloads/Datasets/GDP_currentUS.csv',sep=',',skiprows=4)
```

```python
GDP = GDP.rename(columns={'Country Name': 'Country'})
GDP = GDP.drop(GDP.columns[61:63], axis=1)
GDP = GDP.drop(GDP.columns[1:4], axis=1)
```

```python
GDP.head()
```

```python
GDP_growth = pd.read_csv('/Users/lipinyuan/Downloads/Datasets/GDP growth.csv',sep=',',skiprows=4)
```

```python
GDP_growth = GDP_growth.rename(columns={'Country Name': 'Country'})
GDP_growth = GDP_growth.drop(GDP_growth.columns[61:63], axis=1)
GDP_growth = GDP_growth.drop(GDP_growth.columns[1:4], axis=1)
```

```python
GDP_growth.head()
```

```python
worldcup_1960 = worldcup_new[worldcup_new.Host_Year >= 1960]
```

```python
worldcup_GDP.head()
```
